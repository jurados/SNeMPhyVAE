```python
import os

import numpy as np
import pandas as pd
```

```python
lightcurve_avmk = pd.read_csv('crossmatch_ZTF_ALeRCE_19052024.csv').drop('Unnamed: 0',axis=1)
lightcurve_avmk
```

```python
lightcurve_avmk = lightcurve_avmk[['oid','ra','dec']]
lightcurve_avmk.to_csv('crossmatch_ZTF_ALeRCE_19052024_dust_table.csv',sep=',',index=False)
```

```python
lightcurve_avmk
```

```python
header = pd.read_csv('extinction_tns_procesed.csv',sep='|', nrows=0)
header = pd.read_csv('extinction_tns_procesed.csv',sep='|', nrows=0).drop([header.columns[1],'Unnamed: 0',header.columns[-2]],axis=1).columns
header
```

```python
nasa_infrared_data = pd.read_csv('extinction_tns_procesed.csv',comment='\\',names=header, skiprows=1, delim_whitespace=True)
nasa_infrared_data.columns
concat_dataframe = pd.concat([lightcurve_avmk[['oid','source_id','firstmjd']],nasa_infrared_data['AV_SandF          ']], axis=1)
concat_dataframe['Av_MW'] = concat_dataframe['AV_SandF          ']
concat_dataframe = concat_dataframe.drop('AV_SandF          ', axis=1)
concat_dataframe
```

```python
concat_dataframe.to_pickle('object_ZTF_ALeRCE_19052024.pkl')
```
